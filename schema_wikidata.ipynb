```python
gp = rdflib.Graph()
```

```python
gp.parse('http://www.wikidata.org/entity/Q1708503.ttl')
```

```python
print(len(gp))
```

```python
print(gp.serialize(format = 'turtle').decode())
```

```python
res_list = []

for i, (sub, prop, obj) in enumerate(gp):    
    try:
        nsshort, _, lname = NSM.compute_qname_strict(sub)
        str1 = nsshort + ':' + lname

    except ValueError:
        str1 = sub
        
    try:
        nsshort1, _, lname1 = NSM.compute_qname_strict(prop)    
        str2 = nsshort1 + ':' + lname1
                
    except ValueError:
        str2 = prop
    
    try:
        nsshort2, _, lname2 = NSM.compute_qname_strict(obj)
        str3 = nsshort2 + ':' + lname2
        
    except ValueError:
        str3 = obj
    
    list_n = [i, str1, str2, str3]
    res_list.append(list_n)
    
    if(i == len(gp)):
        break
    
print(res_list)
```

```python
with open('mapping_wikidata.csv', 'w') as f:
    writer = csv.writer(f)
    for row in res_list:
        writer.writerow(row)
```
